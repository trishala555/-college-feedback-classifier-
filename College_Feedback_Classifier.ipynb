```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df['feedback'], df['category'], test_size=0.2, random_state=42)

# Convert text to numbers
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Predict
predictions = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, predictions))
```

```python
import pandas as pd
import random

academics = [
    "Teachers explain concepts clearly",
    "Professors are very knowledgeable",
    "Syllabus is well structured",
    "Teaching methods are interactive",
    "Faculty members are supportive",
    "Some teachers rush through topics",
    "Assignments help in understanding",
    "Lectures are easy to understand",
    "Course content is relevant",
    "Practical sessions are helpful"
]

facilities = [
    "Labs need better equipment",
    "Classrooms are clean and spacious",
    "WiFi is not working properly",
    "Library has good collection of books",
    "Computer labs are outdated",
    "Canteen food quality is poor",
    "Sports facilities are excellent",
    "Hostel rooms need maintenance",
    "Projectors are not functioning",
    "Campus is well maintained"
]

administration = [
    "Office staff is very helpful",
    "Administration is slow in processing requests",
    "Exam process is smooth",
    "Management responds quickly",
    "Accounts department delays payments",
    "Staff behavior is professional",
    "Scholarship process is complicated",
    "Timetable updates are delayed",
    "Communication from office is unclear",
    "Registration process is simple"
]

data = []

for _ in range(35):
    data.append([random.choice(academics), "Academics"])
    data.append([random.choice(facilities), "Facilities"])
    data.append([random.choice(administration), "Administration"])

df = pd.DataFrame(data, columns=["feedback", "category"])
df.to_csv("feedback.csv", index=False)

print("Dataset size:", df.shape)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df['feedback'], df['category'], test_size=0.2, random_state=42)

# Improved TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Predict
predictions = model.predict(X_test_vec)

print("Improved Accuracy:", accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

```python
import pickle

# Save model
pickle.dump(model, open("feedback_model.pkl", "wb"))

# Save vectorizer
pickle.dump(vectorizer, open("tfidf_vectorizer.pkl", "wb"))

print("Model and vectorizer saved successfully!")
```
